# aggregate

```python
Aggregating functions are the ones that reduce the dimension of the returned objects.
Some common aggregating functions are tabulated below:

eg:Function Description

mean() Compute mean of groups

sum() Compute sum of group values

size() Compute group sizes

count() Compute count of group

std() Standard deviation of groups

var() Compute variance of groups

sem() Standard error of the mean of groups

describe() Generates descriptive statistics

first() Compute first of group values

last() Compute last of group values

nth() Take nth value, or a subset if n is a list

min() Compute min of group values

max() Compute max of group values
```

```python
data.groupby('Platform')[['NA_Sales','Global_Sales']]. \
aggregate([np.sum,np.max,np.min])
```

```python
# lamba
data.groupby('Platform')[['NA_Sales','Global_Sales']].aggregate(
    [lambda x: x.max() - x.min(),
    lambda x: x.median() - x.mean()]
)
```

```python
# 相同的功能如果只用Matplotlib
import matplotlib.pyplot as plt
import pandas as pd

# 取得不同物種的名稱
species = iris['species'].unique()

# 設定顏色對應物種
colors = sns.color_palette("Set1", len(species))

# 根據物種繪製不同顏色的點
for i, sp in enumerate(species):
    species_data = iris[iris['species'] == sp]
    plt.scatter(species_data['sepal_length'], species_data['sepal_width'],
                label=sp, color=colors[i], alpha=0.7)

# 設置標題和軸標籤
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# 顯示圖例
plt.legend(title='Species')

# 顯示圖形
plt.show()
```

置信區間是使用自舉法計算的，對於較大的資料集來說，這可能非常耗時。因此可以禁用它們：

```python
# 置信區間是使用自舉法計算的，對於較大的資料集來說，這可能非常耗時。
# 因此可以禁用它們：
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", errorbar=None,
)
```

另一個不錯的選擇，特別是對於較大的數據，是透過繪製標準差而不是置信區間來表示每個時間點的分佈分佈：

```python
# 另一個不錯的選擇，特別是對於較大的數據，
# 是透過繪製標準差而不是置信區間來表示每個時間點的分佈分佈：
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", errorbar="sd",
)
```

```python
# Showing multiple relationships with facets
# 雖然函數可以同時顯示多個語義變量(資料欄位)，但這樣做並不總是有效。
# 當您確實想了解兩個變數之間的關係如何依賴多個其他變數時該怎麼辦？
# 最好的方法可能是製作多個圖。
# 因為 relplot() 是基於 FacetGrid，所以這很容易做到。
# 為了顯示附加變數的影響，不要將其指派給圖中的語義角色之一，而是使用它來「分面」視覺化。
# 這意味著您創建多個軸並在每個軸上繪製資料子集：
```

```python
tips.head()
```

```python
# 在一張圖中分別使用顏色跟點樣式來代表smoker與time
sns.relplot(
    data=tips,
    x="total_bill", y="tip", hue="smoker", style="time",
)

# ↑ 這張圖雖然分別使用顏色跟樣式來代表不同的資料欄位，但是看起來還是很雜亂

# ↓ 比較合適的方法是將 time 這個欄位分成兩張不同的圖來並列顯示
```

```python
sns.relplot(
    data=tips,
    x="total_bill", y="tip", hue="smoker", col="time",
)
```

```python
#練習題
#使用下面資料，仿造上圖來繪圖
penguins = sns.load_dataset("penguins")
penguins.head()
sns.relplot(
    data = penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    col="island",
    style="sex")
```

```python
# 您也可以透過這種方式顯示兩個變數的影響：一種透過在列上分面，一種透過在行上分面。
# 當您開始在網格上新增更多變數時，您可能需要減少圖形大小。
# 請記住，FacetGrid 的大小由每個面的高度和縱橫比參數化：
sns.relplot(
    data=tips,
    x="total_bill", y="tip", hue="smoker", col="time", row="sex"
)
```
