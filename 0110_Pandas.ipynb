```python
#寫程式計算不同平台的總銷售量

#只算一個的時候  data[data['Platform'] == 'Wii']['Global_Sales'].sum()

#共31個平台
plat = data['Platform'].unique()
for n in plat:
  print(n, " : ", data[data['Platform'] == n]['Global_Sales'].sum())
```
